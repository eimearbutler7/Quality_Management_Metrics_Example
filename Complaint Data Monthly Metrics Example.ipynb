```python
df1 = len(df.loc[df['Status'] == "Closed"]) #count how many within this df are at closed status
df2 = len(df.loc[df['Status'] == "Canceled"]) #count how many within this df are at canceled status
df3 = len(df.loc[df['Status'] == "Review"]) #count how many within this df are at Dispositioned status
df4 = len(df.loc[df['Status'] == "Investigation Completed"]) #count how many within this df are at MRB status
df5 = len(df.loc[df['Status'] == "Pending"]) #count how many within this df are at pending status
df6 = len(df.loc[df['Status'] == "Submitted"]) #count how many within this df are at pending status
```

```python
### 1. New = # of complaints or NCRs created during the month
df_org = df.loc[:, 'Date Originated'] #create df of just date originated column
df_orga = df_org.str.split(n=2, expand=True) #split out date from time and context (PST etc.)
df_orgb = pd.to_datetime(df_orga.loc[:,0]) #turn date string into recognised datetime figure 
df_orgc = (df_orgb.loc[:] >= start_date) & (df_orgb.loc[:] <= end_date) #set boundries for new df
len_df_orgc = len(df_orgc.loc[df_orgc[:] == True]) #establish number of rows in new df
```

```python
###  Number of complaints or NCRs CLOSED during the month
df_clo = df.loc[:, 'Final Complete Date'] #create df of just date originated column
df_cloa = df_clo.str.split(n=2, expand=True) #split out date from time and context (PST etc.)
df_clob = pd.to_datetime(df_cloa.loc[:,0]) #turn date string into recognised datetime figure 
df_cloc = (df_clob.loc[:] >= start_date) & (df_clob.loc[:] <= end_date) #set boundries for new df
len_df_cloc = len(df_cloc.loc[df_cloc[:] == True]) #establish number of rows in new df
```

```python
## How many of those closed were closed within 60 days? 
df15 = df_clob - df_orgb
df_15 = pd.DataFrame({'originated_date': df_orgb, 'final_closed_date': df_clob, 'days_to_close': df15}) #add the days to close back into a data rame with both org date and close date
df_16 = df_15.loc[df_15['days_to_close'] <= '60 days'] #filter for those with 60 days or less in the difference

df_17 = (df_16.loc[:, 'final_closed_date'] >= start_date) & (df_16.loc[:, 'final_closed_date'] <= end_date) #set boundries for new df within the month you are looking at
#df_16.loc['1-569158182',:] #This set will give a booleon result of True where the row is less than 60 days and also closed within the last month
len_df_17 = len(df_17.loc[df_17[:] == True]) #establish number of rows in new df
```

```python
## Rate of total complaints processed 2019 vs those processed on time 2019: 
start_date_CA = '2019-1-1' #DO NOT CHANGE

## Total closed in 2019####
df_closed4a = (df_clob.loc[:] >= start_date_CA) & (df_clob.loc[:] <= end_date) #set boundries for new df
len_df_closed4a = len(df_closed4a.loc[df_closed4a[:] == True]) #establish number of rows in new df
print(" ") #print space
print(len_df_closed4a, "records have been closed between Jan 1st 2019 and", end_date)

## How many of those closed were closed within 60 days in 2019? 
df_17a = (df_16.loc[:, 'final_closed_date'] > start_date_CA) & (df_16.loc[:, 'final_closed_date'] <= end_date) #set boundries for new df within the month you are looking at
#df_16.loc['1-569158182',:] #This set will give a booleon result of True where the row is less than 60 days and also closed within the last month
len_df_17a = len(df_17a.loc[df_17a[:] == True]) #establish number of rows in new df
```

```python
## BACKLOG total? 
#Aim: Total open now minus opened this month plus those closed since month end

##### Total open now ######
filter_list = ['Submitted', 'Pending', 'Review', 'Investigation Completed'] #filter for open records
df_bl1 = df[df.Status.isin(filter_list)] #source: https://www.ritchieng.com/pandas-multi-criteria-filtering/
df_bl1a  = df_bl1.loc[:, 'Date Originated'] #create df of just date originated column
df_bl1b = df_bl1a.str.split(n=2, expand=True) #split out date from time and context (PST etc.)
df_bl1c = pd.to_datetime(df_bl1b.loc[:,0]) #turn date string into recognised datetime figure 
df_bl1d = (df_bl1c.loc[:] > end_date)#set boundries for new df
len_df_bl1d = len(df_bl1d.loc[df_bl1d[:] == True]) #establish number of rows in new df
total_open_month_end = len(df_bl1) -len_df_bl1d

#### minus those opened this month #####
total_open_this_month = len_df_orgc #from section 1
total2 = total_open_month_end - total_open_this_month
#total2

#### plus closed since month end ##### - as these would have been open at the end of the last month

filter_list2 = ['Closed'] #can be used to test if numbers of all status categories add up. #remove all 3 hashtags
df_bl3 = df[df.Status.isin(filter_list2)] 
#print(len(df_bl3), "all Complaints are at closed status")
df_bl_closed_recently1 = df_bl3.loc[:, 'Final Complete Date'] #create df of just date originated column
df_bl_closed_recently2 = df_bl_closed_recently1.str.split(n=2, expand=True) #split out date from time and context (PST etc.)
df_bl_closed_recently3 = pd.to_datetime(df_bl_closed_recently2.loc[:,0]) #turn date string into recognised datetime figure 
df_bl_closed_recently4 = (df_bl_closed_recently3.loc[:] > end_date) #set boundries for new df
len_df_bl_closed_recently4 = len(df_bl_closed_recently4.loc[df_bl_closed_recently4[:] == True]) #establish number of rows in new df
#print(len_df_bl_closed_recently4)

total3 = total2 + len_df_bl_closed_recently4
```
